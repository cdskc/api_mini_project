```python
def intraday(date):
    """Return the intraday range (high - low)"""
    return data_nolabel[date][7] - data_nolabel[date][8]
```

```python
max_value = None
max_date = None

for i in range(len(data_nolabel)):
    value = abs(intraday(i))
    
    if max_value is None or value > max_value:
        max_value = intraday(i)
        max_date = data_nolabel[i][0]

print('Greatest day-over-day change')
print(max_date)
formatted_max_value = lambda x: "{:.2f}".format(x)
result = formatted_decimal(max_value)
print(result)
```

```python
def prev_change(date):
    """Return the intraday range (high - low)"""
    return data_nolabel[date][1] - data_nolabel[date + 1][1]
```

```python
max_value = None
max_date = None

for i in range(len(data_nolabel)-1):
    value = abs(prev_change(i))
    
    if max_value is None or value > max_value:
        max_value = prev_change(i)
        max_date = data_nolabel[i][0]

print('Highest intraday range')
print(max_date)
result = formatted_decimal(max_value)
print(result)
```

```python
mean_vol = sum(volume) / len(volume)
result = formatted_decimal(mean_vol)
print('Mean volume: ', str(int(float(result) * 1000)))
```

```python
sorted_volume = sorted(volume)
n = len(sorted_volume)

if n % 2 == 1:  # Odd number of elements
    median = sorted_volume[n // 2]
else:  # Even number of elements
    middle1 = sorted_volume[n // 2 - 1]
    middle2 = sorted_volume[n // 2]
    median = (middle1 + middle2) / 2

print("Median volume:", str(int(median * 1000)))
```
